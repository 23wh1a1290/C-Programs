```python
def generate_content(topic, grade):
    prompt = f"""
    Create lesson notes for topic: {topic}
    Target Grade: {grade}

    Include:
    - Learning Objectives
    - Explanation
    - Examples
    - Practice Questions
    - Answers
    """

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    outputs = model.generate(
        **inputs,
        max_new_tokens=500,
        temperature=0.7
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def generate_lesson(topic, grade, difficulty="Medium"):

    prompt = f"""
    Create structured lesson notes.

    Topic: {topic}
    Grade: {grade}
    Difficulty: {difficulty}

    Include:
    1. Title
    2. Learning objectives
    3. Explanation
    4. Examples
    5. Practice questions
    6. Answers
    7. Summary
    """

    return generate_text(prompt)
```

```python
def generate_quiz(topic, grade, num_questions=5):

    prompt = f"""
    Generate {num_questions} multiple choice questions.

    Topic: {topic}
    Grade: {grade}

    Each question must have:
    - 4 options
    - Clearly marked correct answer
    """

    return generate_text(prompt)
```

```python
def evaluate_answer(question, student_answer):

    prompt = f"""
    Question: {question}
    Student Answer: {student_answer}

    Evaluate correctness.
    Provide feedback.
    """

    return generate_text(prompt)
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import ParagraphStyle, getSampleStyleSheet
from reportlab.lib import colors
from reportlab.lib.units import inch

def save_as_pdf(text, filename="lesson.pdf"):

    doc = SimpleDocTemplate(filename)
    elements = []
    styles = getSampleStyleSheet()

    for line in text.split("\n"):
        elements.append(Paragraph(line, styles["Normal"]))
        elements.append(Spacer(1, 0.2 * inch))

    doc.build(elements)
    print(f"Saved as {filename}")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
def generate_text(prompt, max_tokens=500):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=0.7
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def educational_agent(topic, grade, difficulty, mode):

    if mode == "Lesson":
        prompt = f"""
        Create structured lesson notes.
        Topic: {topic}
        Grade: {grade}
        Difficulty: {difficulty}

        Include:
        - Title
        - Learning objectives
        - Explanation
        - Examples
        - Practice questions
        - Answers
        - Summary
        """

    elif mode == "Quiz":
        prompt = f"""
        Generate 5 multiple choice questions.
        Topic: {topic}
        Grade: {grade}

        Each question must have:
        - 4 options
        - Clearly marked correct answer
        """

    else:
        prompt = f"""
        Explain the topic clearly.
        Topic: {topic}
        Grade: {grade}
        Difficulty: {difficulty}
        """

    return generate_text(prompt)
```

```python
import gradio as gr

with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown("# 🎓 Educational AI Agent")
    gr.Markdown("Generate Lessons, Quizzes, and Explanations Offline")

    with gr.Row():
        topic = gr.Textbox(label="Topic", placeholder="Enter topic (e.g., Photosynthesis)")
        grade = gr.Textbox(label="Grade", placeholder="Enter grade level")

    difficulty = gr.Dropdown(
        choices=["Easy", "Medium", "Hard"],
        value="Medium",
        label="Difficulty"
    )

    mode = gr.Radio(
        choices=["Lesson", "Quiz", "Explanation"],
        value="Lesson",
        label="Mode"
    )

    generate_btn = gr.Button("Generate Content")

    output = gr.Textbox(label="Generated Output", lines=20)

    generate_btn.click(
        educational_agent,
        inputs=[topic, grade, difficulty, mode],
        outputs=output
    )

demo.launch()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "microsoft/Phi-3-mini-4k-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
def generate_text(prompt, max_tokens=800):

    formatted_prompt = f"<|user|>\n{prompt}\n<|assistant|>"

    inputs = tokenizer(formatted_prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=0.7,
        do_sample=True
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response.split("<|assistant|>")[-1].strip()
```

```python
def educational_agent(topic, grade, difficulty, mode):

    if mode == "Lesson":
        prompt = f"""
Create detailed structured lesson notes.

Topic: {topic}
Grade Level: {grade}
Difficulty: {difficulty}

Use clear headings.
Use bullet points.
Provide 3 examples.
Provide 5 practice questions with answers.
End with a short summary.
"""

    elif mode == "Quiz":
        prompt = f"""
Generate 5 multiple choice questions.

Topic: {topic}
Grade Level: {grade}

Each question must:
- Have 4 options
- Clearly mark the correct answer
"""

    else:
        prompt = f"""
Explain this topic clearly and simply:
Topic: {topic}
Grade Level: {grade}
Difficulty: {difficulty}
"""

    return generate_text(prompt)
```
